## Funciones

La fase $\Phi$:
$$
\Phi=\frac{t-t_0}{T}-\left \lfloor \frac{t-t_0}{T} \right\rfloor
$$

Sobre `dual_sort`:

1. Crea un diccionario $d$ de los arreglos $x$, $y$, que relaciona $x_i\mapsto y_i$
2. Define el arreglo $xs$, que contiene a $x$ ordenado de menor a mayor.
3. Define el arreglo $ys$ como $ys_i=d(xs_i)$
4. Retorna $xs$, $ys$

`String_sum` calcula la *longitud de la cuerda* de los arreglos $x$,$y$. Es decir:
$$
\sum_0^{n-1} \sqrt{(x_{i+1}-x_i)^2+(y_{i+1}-y_i)^2}
$$

```python
Vfloat=np.vectorize(float)
Vround=np.vectorize(round)

def phase(t,t_0,T):
    return ((t-t_0)/T)-(Vround((t-t_0)/T))

def dual_sort(x,y):
    d=dict(zip(x,y))
    sx=sorted(x)
    sy=np.zeros(len(sx))
    for i in range(len(sx)):
        sy[i]=(d[sx[i]])
    return sx,sy

def string_sum(x,y):
    a=0
    for i in range(len(x)-1):
        a+=np.sqrt((x[i+1]-x[i])**2+(y[i+1]-y[i])**2)
    return a
```

### Lectura de datos

```python
start=time.time()

data = open( 'OGLE-BLG-CEP-01.dat', 'r' )
lines = np.array(data.readlines( ))
splitted=np.core.defchararray.rsplit(lines)
HJD_string = np.zeros(len(splitted))
Magnitude_string = np.zeros(len(splitted))
error_string = np.zeros(len(splitted))

for i in range(int(len(splitted))):
    HJD_string[i]=(splitted[i][0])
    Magnitude_string[i]=(splitted[i][1])
    error_string[i]=(splitted[i][2])

HJD=Vfloat(HJD_string)
Mag=Vfloat(Magnitude_string)
err=Vfloat(error_string)

end=time.time()
print(end-start)
```

```python
start=time.time()

p=np.arange(0.1,10,0.002) # Estos serán los períodos a probar
print(len(p)) 
sums=np.zeros(len(p)) # String-Length para cada período

count=0
for i in range(len(p)):
    ph=phase(HJD,max(HJD),p[i])
    sph,sm=dual_sort(ph,Mag)
    sums[i]=string_sum(sph,sm)
    count+=1
    if count%(int(len(p)/10.))==0: #Imprime cada 10% del código
        print(str(int(float(i)*100/len(p))+1)+"%")

end=time.time()
print(end-start)
```

```python
per=p[np.argwhere(sums==min(sums))[0][0]] #El período donde la string_sum da mínimo

fig1, ax = plt.subplots(figsize=(20, 8))
plt.scatter(p,sums,s=1,c='k') # Cada punto es [período, string_sum(Período)]
plt.plot([min(p),max(p)],[min(sums),min(sums)]) 
plt.xlabel("Período",fontsize=20)
plt.axis([min(p),max(p),min(sums)*0.9,max(sums)])
plt.scatter([per],[min(sums)]) # Mínimo.

print(min(sums))
print(per)
```
