# Tr√°nsito b

```python
data = open( 'TRAPPIST-1b.dat', 'r' )
lines = data.readlines( )
splitted = [] 
for i in range(len(lines)):
    splitted.append(lines[i].split())
x=np.zeros(len(lines))
y=np.zeros(len(lines))
err=np.zeros(len(lines))
for i in range(len(splitted)):
    x[i]+=float(splitted[i][0])
    y[i]+=float(splitted[i][1])
    err[i]+=float(splitted[i][2])
```

```python
fig1, ax = plt.subplots(figsize=(20, 7))
plt.errorbar(x,y,yerr=err,fmt='k.',ecolor='k',mew=0.1,ms=0,elinewidth=0.3)
plt.scatter(x[:206],y[:206],c='m',s=15,label='TRAPPIST',edgecolors='face')
plt.scatter(x[207:310],y[207:310],c='b',s=15,label='HCT/HFOSC',edgecolors='face')
plt.scatter(x[311:445],y[311:445],c='r',s=15,label='UKIRT/WFCAM',edgecolors='face')
plt.scatter(x[445:],y[445:],c='g',s=15,label='VLT/HAWKI',edgecolors='face')
#ax.set_axis_bgcolor((0, 0, 0))

plt.legend(loc=4)


plt.rc('text', usetex=True) #Tex eneable
plt.rc('font', family='serif') #Set font to serif-like font.

plt.xlabel(r'Time from mid-transit (day)',fontsize=20)
plt.ylabel(r'Flux',fontsize=20)
plt.title(r'TRAPPIST-1b',fontsize=20)
matplotlib.rcParams.update({'font.size': 15})

plt.axis([ min(x), max(x)*1.05, min(y)*0.99, max(y)*1.001 ]) 
plt.show()
```
